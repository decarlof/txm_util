```python
def execute(cmd):
    popen = subprocess.Popen(cmd, stdout=subprocess.PIPE, universal_newlines=True)
    for stdout_line in iter(popen.stdout.readline, ""):
        yield stdout_line 
    popen.stdout.close()
    return_code = popen.wait()
    if return_code:
        raise subprocess.CalledProcessError(return_code, cmd)
```

```python
for idx, scan in enumerate(df_Centers.index[16:24]):
    t0 = time.time()
    print('Converting Scan {} to UInt16 .. '.format(scan), end='')
    
    algorithm = 'astra_fbp'
    h5fname = os.path.join(data_path, scan)
    recon_path = os.path.join(os.path.dirname(h5fname), os.path.splitext(os.path.basename(h5fname))[0]+ '_full_rec', algorithm)

    recon_files = os.listdir(recon_path)
    fname = os.path.join(recon_path, recon_files[0])
    rec_slices = dxchange.read_tiff_stack(fname, ind=range(len(recon_files)))

    rec_slices = rec_slices - rec_slices.min()
    rec_slices *= 65535/rec_slices.max()

    fname = os.path.join(recon_path, 'recon')
    dxchange.write_tiff_stack(np.asarray(rec_slices, dtype=np.uint16), fname=fname, overwrite=True)
    
    print('Done ({:.0f}) sec'.format(time.time() - t0))
```

```python
def convert_to_UInt16(recon_file, recon_path, scan):
    fname = os.path.join(recon_path, recon_file)

    rec_slice = dxchange.read_tiff(fname)
    if rec_slice.dtype == 'uint16':
        return

    s_ = np.s_[60:-60, 500:-500]
    if int(scan[:3]) <= 15:
        s_ = np.s_[750:-600, 100:-100]

    rec_slice = rec_slice[s_]

    # Convert to uint16
    rec_slice = rec_slice - rec_slice.min()
    rec_slice *= 65535/rec_slice.max()

    dxchange.write_tiff(np.asarray(rec_slice, dtype=np.uint16), fname=fname, overwrite=True)
```
